```python
import geopandas as gpd
import pathlib
```

```python
lowJson = pathlib.Path() / 'usa-low.json'
assert lowJson.exists()
highJson = pathlib.Path() / 'usa-high.json'
assert highJson.exists()
```

```python
gdf = gpd.read_file(highJson)
```

```python
gdf['area'] = gdf.apply(lambda row : gpd.GeoSeries(row['geometry']).area, axis = 1)
```

Add `zlfc` = *zoom level full containment*, This field will indicate the maximum zoom level the user can go up to while still keeping the entire tract in view. Below, we sample a few tracts to get an idea of the relationship between zoom level and area

```python
gdf['zlfc'] = 0
gdf.at[9846, 'zlfc'] = 4.45
gdf.at[10212, 'zlfc'] = 7.08
gdf.at[9937, 'zlfc'] = 5.5
gdf.at[9857, 'zlfc'] = 5.5
gdf.at[9935, 'zlfc'] = 5.38
gdf.at[9936, 'zlfc'] = 5.77
gdf.at[9921, 'zlfc'] = 6.1
gdf.at[9847, 'zlfc'] = 5.33
gdf_short = gdf[["GEOID10", "SF", "CF", "area", "zlfc"]]
gdf_short_sorted = gdf_short.sort_values(by='area', ascending=False);
gdf_short_sorted.head(30)
```
